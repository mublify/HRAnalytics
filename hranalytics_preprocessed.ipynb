#### Dummy Variables:

1 if a certain categorical effect is present, 0 if it is absent

```python
#To ensure that the employees were absent for only ONE reason.

reason_columns=pd.get_dummies(df['Reason for Absence'])
reason_columns
```

```python
#Checking Rows by ensuring the same (no repition of reasons)
#0 means missing value
#1 means single value
#2 or higher means repitition

reason_columns['check']=reason_columns.sum(axis=1)
reason_columns
```

```python
#To check the sum of values in the check column. It must be 700.
reason_columns['check'].sum(axis=0)
```

```python
reason_columns['check'].unique()
```

```python
reason_columns=reason_columns.drop(['check'],axis=1)
reason_columns
```

#### Avoiding Multicolinearity

```python
#Dropping Column 0

reason_columns=pd.get_dummies(df['Reason for Absence'],drop_first=True)
reason_columns
```

#### Grouping the reasons for absence

```python
#Dataframe columns

df.columns.values
```

```python
#Reasons Dummy Columns
reason_columns.columns.values
```

```python
df=df.drop(['Reason for Absence'],axis=1)
df
```

```python
#Grouping the Reasons into similar classes using .loc[] method

reas1=reason_columns.loc[:, 1:14].max(axis=1)
reas2=reason_columns.loc[:, 15:17].max(axis=1)
reas3=reason_columns.loc[:, 18:21].max(axis=1)
reas4=reason_columns.loc[:, 22:].max(axis=1)
```

#### Concatenate the data

```python
df=pd.concat([df,reas1,reas2,reas3,reas4],axis=1)
df
```

#### Assigning Column Names

#### Extracting Day of the Week

```python
df_reason_mod['Date'][699].weekday()
```

```python
#Creating a function
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of the Week']=df_reason_mod['Date'].apply(date_to_weekday)
```
